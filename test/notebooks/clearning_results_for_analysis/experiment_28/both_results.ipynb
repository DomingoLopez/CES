```python
import pandas as pd

df_single = pd.read_csv('inference_results_single_cat_clean.csv', header=0, sep=';')
df_multiple = pd.read_csv('inference_results_multiple_cat_clean.csv', header=0, sep=';')
```

```python
df_single = df_single.rename(columns={"category_1": "category_single"})
df_multiple = df_multiple.rename(columns={"category_1": "category_multiple_1", "category_2": "category_multiple_2"})
```

```python
df_single
```

```python

df_multiple.drop(columns=['category_llava'], inplace=True,errors="ignore")
df_multiple
```

```python
df_def = df_single[['id','cluster','img','category_single']] \
    .merge(
        df_multiple[['img','category_multiple_1', 'category_multiple_2']], 
        on='img', how='inner').filter(['id', 'cluster','img', 'category_single', 'category_multiple_1', 'category_multiple_2']
                                      )
```

```python
df_def
```

```python
df_def.to_csv("inference_results_all.csv", sep=";", index=False)
```
